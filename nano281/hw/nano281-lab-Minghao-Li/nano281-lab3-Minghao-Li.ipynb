```python
import pandas as pd

data = pd.read_csv('cod_xrd_42k.csv')
dt = pd.DataFrame(data)
dt
```

```python
# a type of ordinal encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X = dt.iloc[:,4:-1]
y = dt['bravais_lattice']
le.fit(dt['bravais_lattice'])
dt['bravais_lattice_enc'] = le.transform( dt[['bravais_lattice']])
y_enc = dt['bravais_lattice_enc'] 
```

```python
dt.head()
```

```python
X
```

```python
genre_mappings = {index: label for index, label in 
                  enumerate(le.classes_)}
genre_mappings
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y_enc, train_size=0.30, random_state=42)      
```

```python
score_train_X, score_test_X, score_train_y, score_test_y = train_test_split(X, y_enc, train_size=0.80, random_state=42)      
```

```python
train_y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
norm_train_X = scaler.fit_transform(train_X)
norm_test_X = scaler.transform(test_X)
```

```python
score_scaler = StandardScaler()
norm_score_train_X = score_scaler.fit_transform(score_train_X)
norm_score_test_X = score_scaler.transform(score_test_X)
```

```python
norm_train_X
```

```python
# Q3.5.1
# LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(norm_score_train_X, score_train_y)

```

```python
# Q3.5.2
# QuadraticDiscriminantAnalysis

from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

qda = QuadraticDiscriminantAnalysis(store_covariance=True)
qda.fit(norm_score_train_X, score_train_y)
```

```python
# Q3.5.3
# LogisticRegression

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression( max_iter=10000)
lr.fit(score_train_X, score_train_y)
```

```python
accuracy_lda = lda.score(score_test_X, score_test_y)
accuracy_qda = qda.score(score_test_X, score_test_y)
accuracy_lr = lr.score(score_test_X, score_test_y)

print('''
Accuracy:
\n discriminant analysis:%.4f;
\n quadratic discriminant analysis:%.4f;
\n logistic regression:%.4f
'''
%(accuracy_lda,accuracy_qda,accuracy_lr))
```

```python
def plot_grid_search_results(gs, ylim=None):
    """
    Plots the results of GridSearchCV.
    
    Args:
        gs: A GridSearchCV object.
        ylim: Optional setting for y limits.
    """
    results = pd.DataFrame(gs.cv_results_)
    for c in results.columns:
        # Note that here we are working with just variations in one parameter.
        # So we can automatically find the name of that parameter.
        if c.startswith("param_"):
            x = c
            break
    fig, ax = plt.subplots(figsize=(16, 8))    
    ax = sns.lineplot(x=x, y="mean_train_score", data=results)
    ax = sns.scatterplot(x=x, y="mean_train_score", data=results, marker='x')
    ax = sns.lineplot(x=x, y="mean_test_score", data=results)
    ax = sns.scatterplot(x=x, y="mean_test_score", data=results, marker='o')
    plt.xlabel(x)
    if ylim:
        plt.ylim(ylim)
    ax.legend(["Train", "Test"], loc=2);
    
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
```
